```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv("hyperparam_optimization_results.csv")
df.head()

b_df = df[df['model'].isin(['baseline'])]

r_df = df[df['model'].isin(['rnn'])]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=b_df, x='lr', y='val_acc', hue='freeze')
plt.title('Validation Accuracy vs Learning Rate - Baseline')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=r_df, x='lr', y='val_acc', hue='freeze')
plt.title('Validation Accuracy vs Learning Rate - RNN')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=b_df, x='clip', y='val_f1', hue='freeze')
plt.title('Validation F1 vs Gradient Clipping - Baseline')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=r_df, x='clip', y='val_f1', hue='freeze')
plt.title('Validation F1 vs Gradient Clipping - RNN')
plt.show()
```

```python
pivot = b_df.pivot_table(index='hidden_dim', columns='clip', values='val_f1', aggfunc='mean')
plt.figure(figsize=(8, 6))
sns.heatmap(pivot, annot=True, fmt=".3f", cmap="viridis")
plt.title("Validation F1 Heatmap: Hidden Dim vs Clip - Baseline")
plt.show()

pivot = r_df.pivot_table(index='hidden_dim', columns='clip', values='val_f1', aggfunc='mean')
plt.figure(figsize=(8, 6))
sns.heatmap(pivot, annot=True, fmt=".3f", cmap="viridis")
plt.title("Validation F1 Heatmap: Hidden Dim vs Clip - RNN")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=b_df, x='lr', y='val_f1', hue='freeze', marker='o')
plt.title('Validation F1 vs Learning Rate - Baseline')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(data=r_df, x='lr', y='val_f1', hue='freeze', marker='o')
plt.title('Validation F1 vs Learning Rate - RNN')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=b_df, x='val_f1', y='val_acc', hue='hidden_dim', style='freeze', palette='coolwarm')
plt.title("Validation Accuracy vs F1 Score -Baseline")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=r_df, x='val_f1', y='val_acc', hue='hidden_dim', style='freeze', palette='coolwarm')
plt.title("Validation Accuracy vs F1 Score - RNN")
plt.show()
```
